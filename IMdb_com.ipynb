```python
import time
import requests
from IPython.display import clear_output
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = 'https://www.imdb.com/search/title/?release_date=2021&sort=num_votes,desc&page='
mname = []
year = []
rating = []
genre = []
run_tm = []
metascore = []
votes = []
certficate = []


for i in range(1,200,50):
  time.sleep(3)
  req = requests.get(base_url+str(i)+'&ref_=adv_nxt',headers={'Accept-language':'en-US,en;q=0.5'})
  soup = BeautifulSoup(req.text,'html')
  # find each movies 
  movies = soup.findAll('div',attrs={'class':'lister-item-content'}) 
  for m in movies:

   # movie name
   name=m.find('h3',attrs={'class':'lister-item-header'}).a.text
   mname.append(name)
               
               #  year 
   y=m.find('span',attrs={'class':"lister-item-year text-muted unbold"}).text.replace('(','').replace(')','')  
   year.append(y)

  #rating 
   r=m.find('div',attrs={'class':"inline-block ratings-imdb-rating" }).strong.text
   rating.append(r) 

  #certificate
   c=m.find('span',attrs={'class':"certificate"})
   if c is None:
     certficate.append('not available')
   else:
     certficate.append(c.text)

    # run time 
   rt=m.find('span',attrs={'class':'runtime'})
   if rt is None:
     run_tm.append('not available')
   else:
     run_tm.append(rt.text)

    #metascore
  
   ms=m.find('span',attrs={'class':'metascore'})
   if ms is None:
    metascore.append('not available')
   else:
    metascore.append(ms.text)

   # genre  
   g=m.find('span',attrs={'class':'genre'}).text
   genre.append(g)

  #votes  
   v=m.find('span',attrs={'name':'nv'}).text
   votes.append(v)
```

```python
print(len(mname))
print(len(year))
print(len(rating))
print(len(certficate))
print(len(run_tm))
print(len(metascore))
print(len(genre))
print(len(votes))
```

```python
data = pd.DataFrame({'movie_name': mname,
                     'released_yr': year,
                    'ratings': rating,
                    'certificate': certficate,
                    'run time': run_tm ,
                    'metascore': metascore ,
                     'genre': genre ,
                     'votes': votes})
data
```

```python
data.to_csv('imdb_scraped.csv',index=False)
```
